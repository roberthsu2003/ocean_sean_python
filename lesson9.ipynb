```python
import random

min = 1
max = 99
target = random.randint(min, max)
print("=========猜數字數字遊戲===============\n")

while(True):
    keyin = int(input(f"猜數字範圍{min}~{max}:"))
    if(keyin >= min and keyin <= max):
        print("在範圍內")
        break
    else:
        print("不在範圍內")
        continue

print("遊戲結束!")
```
