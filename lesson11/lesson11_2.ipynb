```python
numbers:tuple = (1, 2, 3, 2, 4, 2, 5)

# count() - 統計元素出現次數
print(numbers.count(2))  # 3

# index() - 查找元素索引
print(numbers.index(3))  # 2

# len() - 獲取長度
print(len(numbers))  # 7

# in 運算符
print(2 in numbers)  # True
```
