```python
#dictionary
codes:dict[str,str] = {"tw":"Taiwan", 
                       "jp":"Japan", 
                       'hk':"Hong Kong"}
print(codes["hk"])
print(codes["jp"])
print(codes["tw"])
```

```python
student:dict[str, any] = {'name': 'Alice', 'age': 20, 'grade': 'A'}
print(student['age'])
print(student.get("name"))
print(student.get("name1"))
print(student.get("name1","No name"))
```

```python
numbers = [3, 1, 4, 1, 5, 9, 2, 6]
for  n in numbers:
    print(n)
```

```python
student:dict[str, any] = {'name': 'Alice', 'age': 20, 'grade': 'A'}
for n in student:
    print(f'key:{n},value:{student[n]}')
```
