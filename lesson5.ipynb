```python
#學生總分為300
#有些學生可以加分5%
#如果加分超過300,就以300分為準
try:
  scores = int(input("請輸入學生分數(最高300分):"))
  if scores > 300:
    raise Exception("分數不可以超過300分")

  is_add = input("學生是否符合加分條件?(y,n)")
  #print(scores,is_add)

  if is_add == 'y':
    scores *= 1.05
    if scores > 300:
      scores = 300

  print(f"學生分數是{round(scores)}")
except Exception as e:
  print(f"出現錯誤:{e}")
```

```python
# 90(含)~100分為「優」
# 80(含)~89分為「甲」
# 70(含)~79分為「乙」
# 60(含)~69分為「丙」
# 0(含)~59分為「丁」

score = int(input("請輸入學生分數:"))
if score >= 90:
  print("優")
else:
  if score >= 80:
    print("甲")
  else:
    if score >= 70:
      print("乙")
    else:
      if score >= 60:
        print("丙")
      else:
        print("丁")
```
